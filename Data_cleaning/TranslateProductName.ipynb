```python
import xlrd
import pandas as pd
import os
from pandas import *
from IPython.display import display
import matplotlib
from googletrans import Translator
```

```python
# load excel file
file_path = os.path.join('sales_sheet.xlsx')
# read the excel file
df  = pd.read_excel(file_path, sheet_name=None, ignore_index=True)

# concatenate all worksheets
cdf = pd.concat(df.values())

# display consolidated data
cdf.head()
```

```python
cdf.count()
```

```python
product_list = cdf['Product Name']

product_list[0]
```

```python
translator = Translator(service_urls=[
        'translate.google.com',
        'translate.google.com.cn',
        ])
```

```python
product_list_en = []
for product in product_list:
    # REINITIALIZE THE API
    translator = Translator()
    translation = translator.translate(product, dest='en')
    product_list_en.append(translation.text)
```

```python
product_list_en
```

```python
cdf['Product_Name_EN']=product_list_en
```

```python
cdf.head()
```

```python
writer = pd.ExcelWriter('Clean_Data_Translated.xlsx')
cdf.to_excel(writer,'Sheet1')
```
